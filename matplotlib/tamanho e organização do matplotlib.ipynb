```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5,50) #Return evenly spaced numbers over a specified interval.
```

```python
y=x**2
```

```python
fig, ax = plt.subplots(figsize=(10,10)) #parametro figsize permite dimensionar o tamanho da figura 
ax.plot(x,y,'bx-', label = 'linear')
ax.plot(x,x**3,'gx-',label = 'x^3')
plt.xlabel("eixo x")
plt.ylabel("eixo y")
plt.title('titulo')
plt.legend()
plt.grid()
```
